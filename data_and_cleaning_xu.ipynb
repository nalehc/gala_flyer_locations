These counts are for every n hours. (What is n?) We want total daily entries.
Now make it that we again have the same keys, but now we have a single value for a single day, which is the total number of passengers that entered through this turnstile on this day.

```python
## what does .first mean?
## reset index helps normalize index

dfs_daily = dfs.groupby(["C/A", "UNIT", "SCP", "STATION", "DATE"]).ENTRIES.first().reset_index()
dfs_daily.head()
```

```python
## shift down values by one row

dfs_daily[["PREV_DATE", "PREV_ENTRIES"]] = (dfs_daily
                                                    .groupby(["C/A", "UNIT", "SCP", "STATION"])["DATE", "ENTRIES"]
                                                    .transform(lambda grp: grp.shift(1)))
dfs_daily.head()
```

```python
# Drop last date. drop N/As in Column of PREV_DATE, Axis=0 means row
dfs_daily.dropna(subset=["PREV_DATE"], axis=0, inplace=True)
```

```python
dfs_daily.insert(5, 'DAILY_ENTRIES',dfs_daily.ENTRIES-dfs_daily.PREV_ENTRIES)
dfs_daily.head()
```
